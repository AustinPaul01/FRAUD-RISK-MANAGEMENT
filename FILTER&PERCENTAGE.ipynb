```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TRANSACTION01.csv')
df
```

```python
df['TD_DTRAN_DATE_TIME'] = pd.to_datetime(df['TD_DTRAN_DATE_TIME'])
```

```python
start_date = '2022-12-11 00:00:00'
end_date = '2022-12-12  11:59:59'

filter_df = df[(df['TD_DTRAN_DATE_TIME'] >= start_date) & (df['TD_DTRAN_DATE_TIME'] <= end_date)]
```

```python
filter_df.to_csv('filtered_dataset.csv', index=False)
filter_df
```

```python
filter_df.dtypes
```

```python
null_counts = filter_df.isnull().sum()
null_counts
```

```python
for column in filter_df.columns:
    # Check the data type of the column
    column_dtype = filter_df[column].dtype

    # Fill missing values based on the data type object as rest were filled
    if column_dtype == 'object':
        filter_df[column] = filter_df[column].fillna('Unknown')

# Verify if all missing values are filled
print(filter_df.isnull().sum())
```

```python
fraud = df[mask]
fraud.tail(10)
```

```python
filter_df['Fraud_Cards(Cardholer)'] = None
filter_df['Valid_Cards(Cardholer)'] = None
filter_df['Tran_results(Cardholer)'] = None
filter_df['Fraud_Cards(Merchant)'] = None
filter_df['Fraud_Users(Merchant)'] = None
filter_df['Fraud_Cards(IP)'] = None
filter_df['Valid_cards(IP)'] = None
filter_df['Fraud_Users(IP)'] = None
```

```python
r = []
q = []
for j in range(0, len(filter_df), 1):
    r.clear()
    for i in range(0, 10, 1):
        u = filter_df.iloc[j, :-1]
        count = fraud.iloc[:, i].value_counts()
        if u[i] in fraud.iloc[:, i].values:
            freq = count[u[i]]
        else:
            freq = 0
        r.append(freq)
    q.append(sum(r)/len(fraud)*10)
    print(j)
```

```python
percentage = q

filter_df['percentage'] = percentage
filter_df.head(10)
```
